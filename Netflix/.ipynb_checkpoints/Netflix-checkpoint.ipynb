```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Netflix.xlsx",engine="openpyxl")
```

```python
df
```

```python
diff_ratings=df["rating"].value_counts()
fig,ax=plt.subplots(1,1,figsize=(10,10))
sns.barplot(x=diff_ratings.index,y=diff_ratings,ax=ax)
plt.xticks(np.arange(len(diff_ratings)),diff_ratings.index,rotation=45)
plt.tight_layout()
plt.ylabel("Amount")
plt.xlabel("Ratings")
plt.show()
```
